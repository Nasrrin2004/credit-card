```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
credit_card=pd.read_csv('/content/creditcard.csv')
```

```python
legit = credit_card[credit_card.Class == 0]
fraud = credit_card[credit_card.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
credit_card.groupby('Class').mean()
```

```python
credit_card.value_counts()
```

```python
legit_sample = legit.sample(n=368)
```

```python
new = pd.concat([legit_sample, fraud], axis=0)
```

```python
new.head()
```

```python
new.tail()
```

```python
new['Class'].value_counts()
```

```python
new.groupby('Class').mean()
```

```python
x = new.drop(columns='Class', axis=1)
y = new['Class']
```

```python
print(x)
```

```python
print(y)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, stratify=y, random_state=2)
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train, y_train)
```

```python
x_train_prediction = model.predict(x_train)
training_data_accuracy = accuracy_score(x_train_prediction, y_train)
```

```python
print('accuracy on training data: ',training_data_accuracy)
```

```python
x_test_prediction = model.predict(x_test)
test_data_accuracy = accuracy_score(x_test_prediction, y_test)
```

```python
print('accuracy on test data: ',test_data_accuracy)
```
